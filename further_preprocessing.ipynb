```python
'''
Let us check the transmission column
'''
import pandas as pd

df = pd.read_csv('First_Step-handled_null_values.csv')


pd.set_option('display.max_rows', None)

# this option of display max rows is set so that pandas can display all value counts irrespetcive of number of unique values

print(df['transmission'].value_counts())

print('\n\n')

df['transmission'] = df['transmission'].str.replace('A/T','Automatic')

df['transmission'] = df['transmission'].str.replace('M/T','Manual')

print(df['transmission'].value_counts())


```

```python
'''
Convert all automatic to 'Automatic' and all manual to 'Manual' and all 'cvt' to 'CVT'
'''

df['transmission'] = df['transmission'].replace(['–', 'F', '2', 'SCHEDULED FOR OR IN PRODUCTION'], 'Unknown')
# Define a function to categorize each transmission value
def categorize_transmission(transmission_text):
  # Check if the data is missing (it's a float 'nan' after replacement)
  if pd.isna(transmission_text):
    return 'Unknown'

  # Convert to lowercase to make the search case-insensitive
  transmission_text = str(transmission_text).lower()

  if 'automatic' in transmission_text or 'a/t' in transmission_text:
    return 'Automatic'
  elif 'manual' in transmission_text or 'm/t' in transmission_text:
    return 'Manual'
  elif 'cvt' in transmission_text:
    return 'CVT'
  else:
    return 'Unknown'

# Apply the function to the entire 'transmission' column
df['transmission'] = df['transmission'].apply(categorize_transmission)

# Now, check the new, clean value counts
print(df['transmission'].value_counts())
  # Check for any remaining null values
```

```python
'''
Let us check the engine column
'''

print(df['engine'].value_counts())
```

```python
'''
Damn, a mess of the highest order. 
Lets just try to take number of litres thats it, oof
'''


df['engine_size_L'] = df['engine'].str.extract(r'(\d\.?\d*)\s?(?:L|Liter)')

print(df['engine_size_L'].value_counts())
# Convert the extracted values to numeric, forcing errors to NaN

missing = df['engine_size_L'].isna().sum()
print("Missing:", missing)   # should give 4009 - 3792 = 217

'''
Let us fill the rest of the missing values as unkown
'''

df['engine_size_L'] = df['engine_size_L'].fillna('Unknown')

print(df['engine_size_L'].isna().sum())  # should give 0

'''
Alright now it gives 0, lets drop the engine column 
'''

df = df.drop(columns=['engine'])


```

```python
'''
We have done some significant preproccessing here, let us check the dataframe again
'''

print(df.info())

print(df.describe())

```

```python
'''
Creating some new columns
'''

'''
Age helps us understand since how long the car has been used and miles per year helps us understand how much the car has been used on an average per year which
can be very significant in determining the price of the car
'''
from datetime import datetime
current_year = datetime.now().year  # e.g., 2025

# Car age
df['car_age'] = current_year - df['model_year']

# Avoid divide by zero for current-year cars
df['car_age'] = df['car_age'].replace(0, 1)

# Miles per year
df['miles_per_year'] = df['mileage'] / df['car_age']

print(df[['model_year','mileage','car_age','miles_per_year']].head(15))
```
