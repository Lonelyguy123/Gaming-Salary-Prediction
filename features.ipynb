```python
'''
let us import the second preprocessed file
'''

import pandas as pd
pd.set_option('display.max_rows', None)


df = pd.read_csv('Second_Step-further_preprocessing.csv')

print(df.head())

print(df['exterior_color'].value_counts())
```

```python
'''
Let us check the color columns
'''

#First,going with exterior color


print(df['exterior_color'].value_counts())

print("\n\n")

'''
So much hosh posh, lets clear it and just keep the basic colors(for both exterior and interior)
'''

special_mappings = {
    # Whites
    'wind chill pearl': 'White',
    'bianco isis': 'White',
    'bianco monocerus': 'White',
    'glacier': 'White',
    'ice': 'White',
    'moonlight cloud': 'White',
    'yulong': 'White',

    # Blacks
    'nero daytona': 'Black',
    'nero noctis': 'Black',
    'onyx': 'Black',
    'caviar': 'Black',
    'obsidian': 'Black',

    # Reds
    'scarlet ember': 'Red',
    'rosso': 'Red',
    'rosso corsa': 'Red',
    'ruby flare pearl': 'Red',
    'maroon': 'Red',
    'ember pearlcoat': 'Red',
    'mango tango pearlcoat': 'Red',
    'go mango!': 'Red',
    'arancio borealis': 'Red',
    'rosso mars': 'Red',

    # Blues
    'blu': 'Blue',
    'blu eleos': 'Blue',
    'stormy sea': 'Blue',

    # Greens
    'dark moss': 'Green',
    'verde': 'Green',
    'gecko pearlcoat': 'Green',

    # Grays / Silvers
    'granite': 'Gray',
    'grigio nimbus': 'Gray',
    'liquid platinum': 'Silver',
    'tempest': 'Gray',
    'lunar rock': 'Gray',
    'dark sapphire': 'Gray',
    'designo magno matte': 'Gray',
    'nightfall mica': 'Gray',
    'chalk': 'Gray',

    # Oranges / Yellows
    'hellayella clearcoat': 'Yellow',
    'sunset drift chromaflair': 'Orange',
    'custom color': 'Orange',  # or Unknown if truly random
    'tan': 'Beige',

    # Pinks
    'pink': 'Pink',

    # Metallic leftovers
    'iridescent pearl tricoat': 'Metallic',
    'c / c': 'Metallic'
}


def exterior_color(color_text):
    color_text = str(color_text).lower()

    if color_text in special_mappings:
        return special_mappings[color_text]

    if 'red' in color_text:
        return 'Red'
    elif 'blue' in color_text:
        return 'Blue'
    elif 'black' in color_text:
        return 'Black'
    elif 'white' in color_text:
        return 'White'
    elif 'silver' in color_text:
        return 'Silver'
    elif 'gray' in color_text or 'grey' in color_text:
        return 'Gray'
    elif 'green' in color_text:
        return 'Green'
    elif 'yellow' in color_text:
        return 'Yellow'
    elif 'brown' in color_text:
        return 'Brown'
    elif 'gold' in color_text:
        return 'Gold'
    elif 'orange' in color_text:
        return 'Orange'
    elif 'beige' in color_text:
        return 'Beige'
    elif 'purple' in color_text:
        return 'Purple'
    elif 'pink' in color_text:
        return 'Pink'
    elif 'metallic' in color_text:
        return 'Metallic' 
    else:
        return 'Unknown'  # Keep original if no match

# Apply
df['exterior_color'] = df['exterior_color'].apply(exterior_color)

print(df['exterior_color'].value_counts())

'''
Fixed it,
no hosh posh now, just clean colors
'''


```
